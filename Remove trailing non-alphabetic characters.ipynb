```python
import pandas as pd
import numpy as np
from glob import glob

pd.options.display.max_rows = 200
```

```python
countries = set()

for f in glob('./csvs/*.csv'):
    df = pd.read_csv(f)
    countries |= set(df['country'].tolist())
```

```python
countries_df = pd.DataFrame({ 'country': list(countries) }).sort_values(by='country')
```

```python
countries_df
```

```python
countries_df[countries_df['country'].str.contains(r'\s*[^A-Za-z\s]+$')]
```

```python
replaced_df = countries_df.copy()
replaced_df['country'] = replaced_df['country'].str.replace(r'\s*[^A-Za-z\s]*$', '')
```

```python
replaced_df[countries_df['country'].str.contains(r'\s*[^A-Za-z\s]+$')]
```

Are there any countries with leading/trailing spaces left?

```python
replaced_df[replaced_df['country'].str.contains(r'^\s|\s$')]
```

Countries that appear to be duplicates because they had a trailing space:

```python
replaced_df[replaced_df['country'].duplicated()]
```
