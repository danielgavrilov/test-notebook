```python
import requests
import pandas as pd
```

```python
metadata_groups = requests.get("http://fenixservices.fao.org/faostat/api/v1/en/groupsanddomains?section=download").json()['data']
```

This basically gets the datasets for which there is metadata from the API (there isn't for all!). Just using pandas to display it nicely:

```python
pd.DataFrame(metadata_groups)
```

Get in-depth metadata for each dataset:

```python
def getMetadata(domain_code):
    url = "http://fenixservices.fao.org/faostat/api/v1/en/metadata/%s" % domain_code
    return requests.get(url).json()['data']
```

```python
metadata_by_name = { group['domain_name']: getMetadata(group['domain_code']) for group in metadata_groups }
```

```python
def find(fn, lst):
    for item in lst:
        if fn(item):
            return item
        
```

```python
def get_add_info(metadata):
    return find(lambda row: row['metadata_label'] == 'Statistical concepts and definitions', metadata)

def get_pub_source(metadata):
    return find(lambda row: row['metadata_label'] == 'Source data', metadata)
```

```python
get_add_info(metadata_by_name["Food Balance Sheets"])
```

```python
get_pub_source(metadata_by_name["Food Balance Sheets"])
```
